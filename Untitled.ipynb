```python
train = train.drop(['No'],axis=1)
# train
cnames = train.columns
cnames[0]
```

```python
train_copy = train.copy()
```

```python
remove_data = []
for i in cnames:
    q75,q25 = np.percentile(train.loc[:,i],[75,25])
    iqr = q75 - q25
    
    min = q25-(iqr*1.5)
    max = q75 + (iqr*1.5)
    
    train_copy = train_copy.drop(train_copy[train_copy.loc[:,i]<min].index)
    train_copy = train_copy.drop(train_copy[train_copy.loc[:,i]>max].index)
```
